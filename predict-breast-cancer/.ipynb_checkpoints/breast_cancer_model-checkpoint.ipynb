```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
import numpy as np
```

```python
cancers = load_breast_cancer()
print(cancers.keys())
df_cancer = pd.DataFrame(np.c_[cancers.data, cancers.target], columns=np.append(cancers.feature_names, ['target']))
df_cancer.head()
```

```python
sns.pairplot(df_cancer, hue='target', vars=['mean radius','mean texture','mean perimeter','mean area','mean smoothness'])
```

```python
sns.countplot(df_cancer.target)
```

```python
sns.scatterplot(x='mean area', y='mean smoothness', hue='target',data=df_cancer)
```

```python
plt.figure(figsize=(20,8))
df_correlation = df_cancer.corr()
sns.heatmap(df_correlation, annot=True)
plt.show()
```

```python
df_cancer.target.sum()
```

```python
X = df_cancer.drop('target', axis='columns')
y = df_cancer.target

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=2)
```

```python
print((y_test.sum()/df_cancer.target.sum())*100, (y_train.sum()/df_cancer.target.sum())*100)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix

clf = SVC(gamma='auto', random_state=1)
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
conf_matrix = confusion_matrix(y_test, y_pred)
conf_matrix
```

```python
sns.heatmap(conf_matrix, annot=True)
```

```python
clf.score(X_test,y_test)
```

## **Improving model performance-1**

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

# "scaler" & "svm" can be any name. But they must be placed in the correct order of processing
pipe = Pipeline([("scaler", StandardScaler()), ("svm", clf)])

pipe.fit(X_train, y_train)
```

```python
# Score on testing data
pipe.score(X_test, y_test)
```

```python
y_pred = pipe.predict(X_test)
```

```python
conf_matrix = confusion_matrix(y_test, y_pred)
sns.heatmap(conf_matrix, annot=True)
```

```python
# Score on training data
pipe.score(X_train,y_train)
```

```python
y_pred_train = pipe.predict(X_train)
cm_train = confusion_matrix(y_train, y_pred_train)
sns.heatmap(cm_train, annot=True)
```

## **Improving model performance-2**

```python
sorted(pipe.get_params().keys())[len(pipe.get_params().keys())-20:]
```

```python
c_values = np.linspace(0.1, 1, 10).tolist()
gamma_values = np.linspace(0.00001, 1, 20)

params = {'svm__C':c_values, 'svm__gamma': gamma_values}

from sklearn.model_selection import GridSearchCV

grid = GridSearchCV(pipe, scoring='accuracy', cv=10, param_grid=params)

grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred1 = grid.predict(X_test)

cm1 = confusion_matrix(y_test, y_pred1)

sns.heatmap(cm1, annot=True)
```

```python
grid.score(X_test, y_test)

# this is better score than previous score
```

```python
grid.score(X_train, y_train)
```

```python
y_pred2 = grid.predict(X_train)

cm2 = confusion_matrix(y_train, y_pred2)

sns.heatmap(cm2, annot=True)
```
